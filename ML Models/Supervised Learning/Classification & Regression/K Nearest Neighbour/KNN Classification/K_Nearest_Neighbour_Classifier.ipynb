## K NEAREST NEIGHBOUR CLASSIFIER:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

### BINARY CLASSIFICATION:

```python
from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=3000, #3000 observations or data points
    n_features=4, #4 input features
    n_redundant=1,
    n_classes=2, #binary target/label
    random_state=89
)
```

```python
X
```

```python
y
```

```python
df = pd.DataFrame(X)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

'''
n_neighbors(k)=5 default

weights-uniform and distance.
if all points weighted equally - uniform
if the weighting the points is inverse of their distance(so closer neighbours have greate influyence)


algorithm- need to find distance from new point to every other points to find 5 nearest neighbors.
These algorithms creates a binary tree.
so, due to decrease in search size bcz of trees Time complexity decreases and optimization increases.
can use:

ball_tree
kd_tree
brute
auto(based on the dataset any parameters are selected.)


p-2 (default and int)
This indicates. euclidean distance.
2-euclidean distance by default
if 1 - manhattan distance.

'''
```

```python
classmodel = KNeighborsClassifier() #all default p=2, k=5, algorithm=auto
#if feastures are randomly distribution in features dimensions euclidean distance is prefereed.
```

```python
classmodel
```

```python
classmodel.fit(X_train, y_train)
```

```python
classmodel.get_params()
```

```python
y_pred = classmodel.predict(X_test)
```

```python
error = np.average(y_test-y_pred)
print(error)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(f"Score: {accuracy_score(y_test, y_pred)}")
print(f"Confusion matrix: \n{confusion_matrix(y_test, y_pred)}")
print(f"Classification Report: \n{classification_report(y_test, y_pred)}")
```

### GRID SEARCH CV:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
params = {
    'n_neighbors': [2,4,5,7,8,9],  # K values
    'weights': ['uniform', 'distance'],  # strategies of weights
    'metric': ['euclidean', 'manhattan', 'minkowski'],  # Distance measures
    'p': [1, 2]  # power parameter ( distance behavior)
}

grid_knc = GridSearchCV(KNeighborsClassifier(), param_grid=params, cv=5, scoring='accuracy', n_jobs=-1)
grid_knc.fit(X_train, y_train)


print(f"Best Parameters: {grid_knc.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_knc.best_score_:.4f}")


best_knn = grid_knc.best_estimator_
test_accuracy = best_knn.score(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```
