# **DIABETES PREDICTIONS:**

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset.DESCR)
```

```python
import pandas as pd

df = pd.DataFrame(data=dataset.data, columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
df
```

```python
X=df
y=dataset['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.28,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,11))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor

dtreg = DecisionTreeRegressor()
dtreg.fit(X_train,y_train)
```

**Hyperparameter tuning:**

```python
param = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,7,9,10,13,15,20,25],
    'max_features':['auto','sqrt', 'log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV

grid_dt = GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=param,cv=5,scoring='neg_mean_squared_error')

```

```python
import warnings
warnings.filterwarnings('ignore')

grid_dt.fit(X_train,y_train)

```

```python
grid_dt.best_params_
```

```python
y_pred=grid_dt.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print(r2_score(y_pred,y_test))
print(mean_absolute_error(y_pred,y_test))
print(mean_squared_error(y_pred,y_test))
```

```python
select_model = DecisionTreeRegressor(criterion='poisson', splitter='best', max_depth=2, max_features='log2')

```

```python
select_model
```

```python
select_model.fit(X_train,y_train)
```

```python
#visualizing the model

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(15,11))
tree.plot_tree(select_model, filled=True)
```
