```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification #create a dataset with classification approiach
#doneed standardizattion as well- already created in such way
```

```python
#create a dataset:
X,y=make_classification(n_samples=1000,n_features=10,n_redundant=3,n_informative=3,n_classes=4,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.28, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
mullogmodel = LogisticRegression(multi_class='ovr')

#can assign ridge and lasso for overfitting and feature selectiuon using penality parameter.
#C- inversly proportional to lambda.
#class weight - useful for imbalancve dataset.

#multi_class-auto,ovr, multinomial
#default -auto
#ovr-one vs rest: based on the number of categories internally that many models are created.
```

```python
mullogmodel
```

```python
mullogmodel.fit(X_train,y_train)
```

```python
y_predmullog = mullogmodel.predict(X_test)
```

```python
y_predmullog
```

```python
mullogmodel.predict_proba(X_test) #here for 4 categories of output getting 4 values as probability
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X_t = pd.DataFrame(X_train)
X_te = pd.DataFrame(X_test)
for column in X_t.columns:
    plt.scatter(X_t[column], y_train, color='blue')
    plt.title(f'Logistic Regression Training Data for {column}')
    plt.show()
```

```python
for column in X_te.columns:
    plt.scatter(X_te[column], y_test, color='green', label='True Values')
    for i in mullogmodel.classes_:
      plt.scatter(X_te[column], sigmoid(X_te[column]), color='k', label=f'Logistic Curve of Class {i}')
    plt.scatter(X_te[column], mullogmodel.predict(X_test), color='red', label='Predicted Values')
    plt.title(f'Logistic Regression Prediction for {column}')
    plt.show()
```

```python
#performance metrics:
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_predmullog)
print(score)
```

```python
cm = confusion_matrix(y_test, y_predmullog)
print(cm)
```

```python
print(classification_report(y_test, y_predmullog))
```

## HYPER PARAMETER TUNING & CROSS VALIDATION:

```python
#1. Grid Search CV
#2.  Randomised Search CV

#selecting diff parameters is possible in hyperparameter tuning.
model = LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
C_values = [100,10,1.0,0.1, 0.05, 1.98]#inverse to lambda
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
#these are main key parameter
#to pass this will have to make it key value pair

params = dict(penalty=penalty, C=C_values, solver=solver)

#now need to play with these to check for best parameters:
params

```

### 1. GRID SEARCH CV:

```python
#with this help will find best parameter that fits for dataset.
from sklearn.model_selection import GridSearchCV, KFold
cv = KFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)

#estimator - model used, param_grid - parameters used, scoring-accuracy in binary classification.
#cv used is Stratified k Fold
# n_jobs = -1 will lets you use all processors in the system.
```

```python
grid
```

```python
#This finds best combination and best parameters that fits the model.
grid.fit(X_train, y_train)
#finds best parameters for each parameter combination.
```

```python
grid.best_params_
```

```python
#these are best parameters that are selected.
grid.best_score_
```

```python
y_predgrid=grid.predict(X_test)
```

```python
y_predgrid
```

```python
score = accuracy_score(y_test, y_predgrid)
print(score)
cm = confusion_matrix(y_test, y_predgrid)
print(cm)
print(classification_report(y_test, y_predgrid))
```

```python
for column in X_te.columns:
    plt.scatter(X_te[column], y_test, color='green', label='True Values')
    for i in mullogmodel.classes_:
      plt.scatter(X_te[column], sigmoid(X_te[column]), color='k', label=f'Logistic Curve of Class {i}')
    plt.scatter(X_te[column], grid.predict(X_test), color='red', label='Predicted Values')
    plt.title(f'Logistic Regression Prediction using Grid Search CV for {column}')
    plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```
